```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#html = 'https://www.kaggle.com/zynicide/wine-reviews/data#winemag-data_first150k.csv'
#pd.read_csv(html)
df = pd.read_csv("../Homework/winemag-data_first150k.csv")
df.head()
```

```python
df.groupby(['country']).count()
```

```python
array = ['US','Mexico','Canada']
df_NA = df.loc[df['country'].isin(array)]
```

```python
import seaborn as sns
sns.set(style="ticks", palette="pastel")
sns.boxplot(x="country", y="points",
            palette=["m", "g"],
            data=df_NA)
sns.despine(offset=10, trim=True)
```
